**W tym dokumencie zawarte są wymagane wykresy :)**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x):
    """
    funkcja zwracająca wymagane dane
    """
    return x**2+5
```

```python
def plot(a,b):
    """funkcja plotująca wykresy z danego przedziału"""
    t = np.arange(a, b, 0.1)

    plt.plot(t, f(t), 'r--')
    plt.title("Wykres")
    plt.legend(["zakres {} do {}".format(a,b)])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
plot(-1,1)
plot(-6,6)
plot(0,5)
```
